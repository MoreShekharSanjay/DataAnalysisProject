```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = None
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv", header='infer')
data.head(2)
```

```python
data = data[data["Attrition"]=="Yes"]
```

```python
data.drop(["Attrition", "EmployeeCount", "EmployeeNumber"], axis=1, inplace=True)
data.head(2)
```

```python
orig_columns = pd.get_dummies(data).columns
# orig_columns
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
X = data.drop(["YearsAtCompany"], axis=1)
y = data["YearsAtCompany"]
```

```python
import statsmodels.api as sm
```

```python
X.columns
```

```python
X2 = sm.add_constant(X.astype(float))
est = sm.OLS(y.astype(float), X2)
est2 = est.fit()
print(est2.summary())
```

```python
keep = ["JobSatisfaction", "NumCompaniesWorked", "TotalWorkingYears", "TrainingTimesLastYear", "YearsInCurrentRole", 
        "YearsSinceLastPromotion", "YearsWithCurrManager", "Department_Research & Development", "Department_Sales",
        "JobRole_Manager", "JobRole_Sales Executive", "JobRole_Sales Representative", "OverTime_Yes"]
```

```python
X_kept = X.loc[:, keep]
```

```python
X_kept.head()
```
